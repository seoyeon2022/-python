```python
#202201722 서연
#구간 [0,2]에서 함수 f(x) = x^2 + sin(3x+7) 함수의 그래프를 그리세요. 

import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(0,2)
b = a**2 + np.sin(3*a + 7)

plt.plot(a,b)
plt.show()
```
